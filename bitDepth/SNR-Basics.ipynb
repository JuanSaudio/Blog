```python
# nbi:hide_in
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
n = np.arange(N)
f = 5
x = np.sin(2 * np.pi * f * n / N)
y = np.random.normal(size=N) * 0.01
```

```python
plt.plot(n, y)
plt.plot(n, x)
plt.grid(True)
plt.autoscale(tight=True)
plt.annotate('SNR?', xy=(5900, 0.504), xytext=(5200, 0.46), ha='center', va='bottom', 
             arrowprops=dict(arrowstyle='-[, widthB=5, lengthB=1', lw=1.0))
```
